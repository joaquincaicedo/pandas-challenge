```python
import pandas as pd
import numpy as np
```

```python
# File to Load
file_to_load = "purchase_data.csv"
# Read Purchasing File and store into Pandas data frame
data = pd.read_csv(file_to_load)
data.head(5)
```

```python
#Create Total PLayer
players = len(data['SN'].value_counts())
total_players = {'Total Players' : [players] }
total_players = pd.DataFrame(total_players)
total_players
```

```python
items = len(data['Item ID'].value_counts())
avg_price = data['Price'].mean()
purchases = data['Price'].count()
revenue = data['Price'].sum()
sales_dict = {'Items' : [items], 'Average Price' : [avg_price], 'Purchases' : [purchases], 'Revenue' : [revenue] }
sales_data = pd.DataFrame(sales_dict)
sales_data
```

```python
SN_count = data.loc[:,['SN', 'Gender']]
SN_count = SN_count.drop_duplicates()
gender_count = pd.DataFrame(SN_count['Gender'].value_counts())
gender_count['Gender Ratio'] = (gender_count['Gender'] / players).map("{:,.1f}%".format) 
gender_count
```

```python
new_data = pd.DataFrame(data.groupby('Gender')['Purchase ID'].count())
new_data['Avg Purchase Price'] = pd.DataFrame(data.groupby('Gender')['Price'].mean())
new_data['total Purchase'] = pd.DataFrame(data.groupby('Gender')['Price'].sum())
new_data['Avg Purchase per Person'] = new_data['total Purchase'] / gender_count['Gender']
new_data
```

```python
age_data = data.drop_duplicates("SN", keep="first")
bins = pd.cut(age_data['Age'], [0,9, 14,19,24,29,34,39,200])
age_data_df = pd.DataFrame(age_data.groupby(bins)['Age'].count())
age_data_df = age_data_df.rename(columns={"Age":"Total Count"})
age_data_df['Percentage of Player'] = age_data_df['Total Count'] / age_data_df['Total Count'].sum()
age_data_df
```
